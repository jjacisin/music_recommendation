## Computing Popularity 

```python
f = open('kaggle_visible_evaluation_triplets.txt')
lines = f.readlines()
```

```python
lines[0]
```

```python
lines[0].strip()
```

```python
lines[0].strip().split('\t')
```

```python
def song_parser(data):
    local_dict = dict()
    for line in data:
        _, song, _ = line.strip().split('\t')
        if song in local_dict:
            local_dict[song] += 1
        else:
            local_dict[song] = 1
    return local_dict
```

```python
song_to_count = song_parser(lines)
```

```python
song_to_count
```

## Song Ordered Descending

```python
song_ordered = sorted(song_to_count, key=lambda song:song_to_count[song],reverse=True)
```
